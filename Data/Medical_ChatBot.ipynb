```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(DATA_PATH)
print(df.shape)
df.head()
df.info()
df.describe(include='all')
```

```python
print(df.isnull().sum())

df = df.dropna()

print(df.shape)
print(df.isnull().sum())
```

```python
print("Duplicate rows:", df.duplicated().sum())
```

```python
df = df.drop_duplicates()

print(df.shape)
print("Duplicate rows:", df.duplicated().sum())
```

```python
print(df['question'].head(10))

df['question'] = df['question'].str.lower()
df['answer'] = df['answer'].str.lower()

df['question'] = df['question'].str.strip().str.replace(r'\s+', ' ', regex=True)
df['answer'] = df['answer'] = df['answer'].str.strip().str.replace(r'\s+', ' ', regex=True)

print(df['question'].head(10))
```

```python
import re

def clean_text(text):
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)   # Remove punctuation
    text = re.sub(r'\s+', ' ', text)             # Remove extra spaces
    return text.strip()

df['question'] = df['question'].apply(clean_text)
df['answer'] = df['answer'].apply(clean_text)
print(df['question'].head(10))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()

question_vectors = vectorizer.fit_transform(df['question'])

print(question_vectors.shape)
```

```python
print(vectorizer.get_feature_names_out()[:30])

print(question_vectors[0])
```

```python
feature_names = vectorizer.get_feature_names_out()

indices = question_vectors[0].indices
scores = question_vectors[0].data

for index, score in zip(indices, scores):
    print(feature_names[index], ":", score)
```

```python
df.to_csv('/content/drive/MyDrive/Medical_ChatBot/medquad_cleaned.csv', index=False)

print("Preprocessed dataset saved successfully!")
```
